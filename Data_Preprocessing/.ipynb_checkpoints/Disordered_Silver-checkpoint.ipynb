```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../datasets/Gold/Au_nanoparticle_dataset.csv")
df
```

```python
df = df.dropna()
df
```

```python
df_subset = df.drop(['ID', 'Formation_E', 'Total_E'], axis=1)
df_subset
```

```python
# df_subset = pd.get_dummies(df_subset)
# df_subset
```

```python
data = df_subset
labels_true = df[['Formation_E']]
scaler = StandardScaler()

for i in range(20):
    X_train, X_test, y_train, y_test = train_test_split(data, labels_true, test_size=0.2)
    X_train = scaler.fit_transform(X_train.values.astype(np.float32))
    X_test = scaler.transform(X_test.values.astype(np.float32))
    
    np.save("..\datasets\Gold_Nano\\X_train" + str(i) +".npy", X_train)
    np.save("..\datasets\Gold_Nano\\X_test" + str(i) +".npy", X_test)
    np.save("..\datasets\Gold_Nano\\y_train" + str(i) +".npy", y_train)
    np.save("..\datasets\Gold_Nano\\y_test" + str(i) +".npy", y_test)
```

```python
from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import pandas as pd

total_test_r2 = []
total_test_mse = []

iteration = 20

seed_rf = np.load(file="..\datasets\Gold_Nano\seed_for_RF.npy")

for iters in range(iteration):
    print("Iteration:", iters)
    name1 = "..\datasets\Gold_Nano\X_train" + str(iters) + ".npy"
    name2 = "..\datasets\Gold_Nano\X_test" + str(iters) + ".npy"
    name3 = "..\datasets\Gold_Nano\y_train" + str(iters) + ".npy"
    name4 = "..\datasets\Gold_Nano\y_test" + str(iters) + ".npy"
    
    X_train = np.load(name1, allow_pickle=True).astype(np.float32)
    print(X_train.shape)
    X_test = np.load(name2, allow_pickle=True).astype(np.float32)
    y_train = np.load(name3, allow_pickle=True).astype(np.float32).reshape(-1, 1)
    y_test = np.load(name4, allow_pickle=True).astype(np.float32).reshape(-1, 1)
    # print(X_train.shape)
    # print(X_test.shape)
    # print(y_train.shape)
    # print(y_test.shape)
    
    estimator=RandomForestRegressor(n_estimators=100, random_state=seed_rf[iters])
    estimator.fit(X_train, y_train.ravel())
    prediction_=estimator.predict(X_test)
    
    total_test_r2.append(r2_score(y_test, prediction_))
    total_test_mse.append(mean_squared_error(y_test, prediction_))
```

```python
print("Median R2:", np.mean(np.array(total_test_r2)))
print("Median RMSE:", np.mean(np.sqrt(np.array(total_test_mse))))
```
